```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(time):
    return np.sin(0.7*time) + 5

def f_d(time):
    return 0.7 * np.cos(0.7*time)
```

```python
def diff(y, time_del):
    y_d = []
    for i in range(len(y)-1):
        
        y_d.append((y[i+1] - y[i]) / time_del)
    
    return y_d

# HW - Try the the diff with noisy y
```

```python
time = np.linspace(0, 5, 1000)
y = f(time=time)
# y_d = f_d(time=time)

y_d_num = diff(y=y, time_del=time[1] - time[0])

plt.plot(time, y, label="y")
# plt.plot(time, y_d, label="y_d")
plt.plot(time[:-1], y_d_num, label="y_d_num")
plt.grid()
plt.legend()
```

# Integration

```python
def integrate(velocity, time_del, ic=0):
    area = [ic]  # It is the starting position, in reality 0 will be the intial condition 
    for i in range(len(velocity)-1):
        trapezoid_area = time_del * (velocity[i] + velocity[i+1]) / 2
        area.append(area[-1] + trapezoid_area)

    return area
```

```python
time = np.linspace(0, 5, 1000)
velocity = f_d(time=time)

time_del = time[1] - time[0]

position = integrate(velocity=velocity, time_del=time_del, ic=0)  # Velocity integral

plt.plot(time, velocity, label="velocity", color="green")
plt.plot(time, position, label="position", color="blue")
# plt.plot(time, y_d, label="y_d")
# plt.plot(time[:-1], y_d_num, label="y_d_num")
plt.grid()
plt.legend()
```
